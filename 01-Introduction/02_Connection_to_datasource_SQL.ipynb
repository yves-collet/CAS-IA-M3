# Connection to SQL Datasource

The following command creates a connection to a SQL Server datasource if it does not already exist:

```sql
%sql
DROP CATALOG IF EXISTS yco_adventureworks;
DROP CONNECTION IF EXISTS yco_connection_to_sql_datasource;
```

```sql
%sql
CREATE CONNECTION IF NOT EXISTS yco_connection_to_sql_datasource TYPE sqlserver
OPTIONS (
  host 'sql-yvco-datasource-prod-001.database.windows.net',
  port '1433',
  user 'yco-admin',
  password 'cas-ia2025'
);
```

This allows us to create a FOREIGN CATALOG to browse the sqldb in Azure without importing the data.

```sql
%sql
CREATE FOREIGN CATALOG IF NOT EXISTS yco_adventureworks USING CONNECTION yco_connection_to_sql_datasource
OPTIONS (database 'sqldb-yco-adventureworks-prod-001');
```
